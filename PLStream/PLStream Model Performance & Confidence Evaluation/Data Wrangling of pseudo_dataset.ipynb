`pseudo_dataset` produced by the Python script `plstream_linear.py` contains tweet sentiment predictions generated by the model & cosine similarity difference values in the form of a list. 

**Data Format:** [index, |cosine similarity difference|, predicted label, {neg_coefficient, pos_coefficient, true_label, 'cos_sim_bad', 'cos_sim_good'}]

This Jupyter notebook contains data wrangling code I wrote to convert & export this data in list format to a csv file.

```python
def predicted(a):
  return a[2]

pred_list = map(predicted, pseudo_dataset)
```

```python
pred_list = list(pred_list)
type(pred_list)
```

```python
def csd(a):
  return a[1]

csd_list = map(csd, pseudo_dataset)
csd_list = list(csd_list)
```

```python
def true_label(a):
  pseudo_dict = a[4]
  return pseudo_dict['true_label']

truth_list = map(true_label, pseudo_dataset)
truth_list = list(truth_list)
```

**Converting lists into dataframe columns**

```python
plot_data = pd.DataFrame(list(zip(pred_list, truth_list, csd_list)),
               columns=['predictions', 'truth', 'cosine_similarity_difference'])
```

```python
plot_data['correct_guess'] = np.where(plot_data['predictions'] == plot_data['truth'], 1, 0)
```

```python
plot_data.shape
```

**Exporting the final dataframe as csv file**

```python
plot_data.to_csv('pseudo_dataset.csv', sep='\t', encoding='utf-8', index = False)
```
